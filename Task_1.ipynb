2. If-Else

```python
n = int(input())
if n % 2 == 1:
    print("Weird")
elif 2 <= n <= 5:
    print("Not Weird")
elif 6 <= n <= 20:
    print("Weird")
elif n > 20:
    print("Not Weird")
```

6. Write a Function

```python
def is_leap(year):
    leap = False
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
      leap = True
    return leap
year = int(input())
print(is_leap(year))
```
